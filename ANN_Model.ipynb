```python
# Import Required libraries --
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
# Load Dataset --
import numpy as np
(x_train, y_train), (x_test, y_test) = mnist.load_data()
#print(np.unique(x_train))
#print(np.unique(y_train))
data1 = np.unique(x_train)
data2 = np.unique(y_train)
print(data1)
print(data2)
```

```python
# Normalize dataset --
x_train = x_train/255.0
x_test = x_test/255.0
print(np.unique(x_train))
```

```python
for i in range(10):
  plt.imshow(x_train[i],cmap="gray")

```

```python
# create ANN model
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation="relu"),
    Dense(128,activation="relu"),
    Dense(10,activation="softmax")
])
```

```python
# Compile the model --
model.compile(optimizer="adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
# train amodel --
model.fit(x_train,y_train,epochs=10)
```

```python
pre = model.predict(x_test)
print(np.unique(pre))
print(pre)
```

```python
pre_label = np.argmax(pre)
print(np.unique(pre_label))
print(pre_label)
```

```python
plt.figure(figsize=(10,8))
for i in range(10):
  plt.subplot(2,5,i+1)
  plt.imshow(x_test[i],cmap="gray")
  pre_label = np.argmax(pre[i])
  actual = y_test[i]
  color = "green" if pre_label == actual else"red"
  plt.title(f"pre: {pre}\n actual:{actual}",color = color)
  plt.axis("off")
plt.tight_layout()
plt.show()
#plt.xticks([])
#plt.yticks([])
  #plt.imshow(x_test[i],cmap="gray")
```
