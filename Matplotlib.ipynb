```python
font_dic={'color':'white','fontfamily':'monospace','fontsize':20,'fontweight':'extra bold'}
```

```python
fig = plt.figure(facecolor='red')
ax = fig.add_axes([0, 0, 1, 1], facecolor='black')
x = np.linspace(-5, 5, 10)

ax.plot(x, x**2, 'r--o', label='square')
ax.plot(x, x**3, 'w:s', label='cube')
ax.legend()
ax.set_title("Anatomy of figure", fontsize=30, fontdict=font_dic)
ax.set_xlabel("x-axis", fontsize=20, fontdict=font_dic)
ax.set_ylabel("y-axis", fontsize=20, fontdict=font_dic)
ax.set_xticks(np.arange(-6,6))
ax.set_xticklabels([chr(65+i) for i in range(12)], rotation=45,fontdict=font_dic)
ax.set_yticks(np.arange(-130, 130, 20))
#xmin, xman, ymin, ymax
ax.grid()
plt.show()

```

```python
"""Object Oriented Method for creating Graphs - this will be useful for complex analytical or analysis application"""
a = [10,20,45,55]
b, = plt.plot(a) #plot() will return a tuple
print("type of b is: ", type(b)) #may vary based on data
b.set_color('r')
plt.show()
print(type(plt.plot()))
```

```python
"""plt.figure() function returns Figure
* When we plot data on a  graph, that graph will be drawn on a figure.
* By default, a figure is automatically generated during ploting in matplotlib.
"""
fig = plt.figure() #figure object created,it is kind of blank canvas.
x = numpy.linspace(0,5,11)
y = x**2
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9]) #4 parameter provided in a list-left, bottom, width, height
plt.plot(x,y)
```

```python
#Change resolution using dpi
plt.figure()
a = 5
b = -4
c = -10
x = np.linspace(-10, 10, num = 100)
y = a*x**3 + b*x + c + b*x**2
plt.grid()
plt.title("With DPI")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.show()
```

```python
ax.plot(x, x**2, x, x**3, lw=2)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18)
ax.set_title("Custom Ticks")
```
